```python
# Correlation matrix
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna('columns') # drop columns with NaN
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, cmap = "Blues", fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix', fontsize=15)
    plt.show()
```

```python
df_all_param = data_f[["Hour", "Temperature(°C)", "Humidity(%)", "Wind_speed_(m/s)", "Visibility_(10m)", "Solar_Radiation_(MJ/m2)", 
             "Rainfall(mm)", "Snowfall_(cm)", "Seasons", "Holiday"]]
df_all_param.head()
```

```python
#Choix du dataframe
X = df_all_param
y = data_f["Rented_Bike_Count"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
print("Train set : ", X_train.shape)
print("Test set : ", X_test.shape)
```

```python
#Modelisation
from sklearn.linear_model import Lasso
```

```python
lasso_reg = Lasso(alpha=0.1)
#training
lasso_reg.fit(X_train, y_train)
#prédiction
prediction = lasso_reg.predict(X_test)
prediction
```

```python
lasso_reg.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Cross-validation with GridSearchCV.
#let's try 30 differents alphas and see which combination is the best
lasso = Lasso(random_state=0, max_iter=10000)
alphas = np.logspace(-3, -0.3, 30)
#This numpy function means that we will test 30 differents alphas between 10^-3 and 10^-0.3 logarithmically
tuned_parameters = [{'alpha': alphas}]
n_folds = 5

gridCV = GridSearchCV(lasso, tuned_parameters, cv=n_folds)
gridCV.fit(X_train, y_train)
```

```python
scores = gridCV.cv_results_['mean_test_score']
plt.figure().set_size_inches(8, 6)
plt.ylabel("score")
plt.xlabel("alpha")
plt.semilogx(alphas, scores)
```

```python
gridCV.best_score_
```

```python
gridCV.best_estimator_
```

```python
#We saw that the best model according to our cross-validation is when our hyper-parameter is alpha=0.3264
```

```python
# ElasticNet
from sklearn.linear_model import ElasticNet
#This method is similar to LASSO but add a coefficient which depends on l1_ratio. If l1_ratio=1 we have the LASSO's equation 
enet = ElasticNet(alpha=0.1, l1_ratio=0.5)

y_pred_enet = enet.fit(X_train, y_train).predict(X_test)
print(enet)
print("r^2 on test data : %f" % enet.score(X_test, y_test))
```

```python
#Cross-validation with GridSearchCV.
#let's try 30 differents alphas and 20 differets l1_ratios and see which combination is the best

alphas2 = np.logspace(-4, -0.3, 30)
#This numpy function means that we will test 30 differents alphas between 10^-4 and 10^-0.5 logarithmically
l1_ratios = np.linspace(0, 1, 20)
#same as above but linearly. Here 20 values between 0 and 1.

tuned_parameters2 = [{'alpha': alphas2, 'l1_ratio': l1_ratios}]
n_folds = 5

gridCV2 = GridSearchCV(ElasticNet(), tuned_parameters2, cv=n_folds)
gridCV2.fit(X_train, y_train)
```

```python
gridCV2.best_estimator_
```

```python
gridCV2.best_score_
```

```python
# SVR
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr_reg.fit(X_train, y_train)
svr_pred = svr_reg.predict(X_test)
svr_pred
```

```python
svr_reg.score(X_test, y_test)
```

```python
#Code in order to make ElasticNet params printable in our last tab
p1 = gridCV2.best_params_['alpha'].round(5)
p2 = gridCV2.best_params_['l1_ratio'].round(5)
gridCV2.best_params_ = {'alpha' : p1, 'l1_ratio' : p2}
gridCV2.best_params_
```

```python
# Display results

scores = []
scores.append({'model': "Lasso", 'score' : gridCV.best_score_, 'best parameters' : gridCV.best_params_})
scores.append({'model': "ElasticNet", 'score' : gridCV2.best_score_, 'best parameters' : gridCV2.best_params_})
scores.append({'model': "SVR", 'score' : svr_reg.score(X_test, y_test)})
scores
```

```python
tab = pd.DataFrame(scores, columns =['model', 'score', 'best parameters'])
tab
```

```python
#Lasso and ElasticNet have very similar results. Those scores are the coefficient of determination R² of our predictions.
#R² is defined as (1-u/v) where u is the RSS and v the TSS (Residual and Total Sum of Squares)
```
